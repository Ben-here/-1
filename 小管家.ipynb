```python
import pandas as pd
import random
import re
import sys

def accounting():
    print("\n好啊，那我們來記帳吧！")
    try:
        df = pd.read_csv('accounting.csv')
        print("已載入現有記帳資料.")
    except FileNotFoundError:
        df = pd.DataFrame(columns=['日期', '類別', '金額', '說明'])
        print("建立新的記帳表.")

    while True:
        action = input("那麼你要做什麼(記一筆/查詢/返回): ")
        if action == '1' or action == '記一筆':
            while True:
                date = input("請輸入日期 (例如: YYYY-MM-DD): ")
                if re.match(r'\d{4}-\d{2}-\d{2}', date):
                    break
                else:
                    print("嗯~~~日期格式好像不太正確，穿到哪個異世界去啦!")
                    print("請輸入 YYYY-MM-DD 格式.")

            while True:
                category = input("請輸入類別 (花錢/進帳): ")
                if category in ['花錢', '進帳']:
                    break
                else:
                    print("阿不就兩個選項，'花錢' 和 '進帳'，難不成還有其他選項.")

            if category == '花錢':
                amount = input("來吧，花了多少: ")
            else:
                amount = input("好棒，那麼賺了多少，不要一塊兩塊的記喔!: ")

            if category == '花錢':
                description = input("那麼，花到哪裡了，好好說說吧:")
            else:
                description = input("那麼，錢怎麼來的，最好不是非法: ")

            new_entry = pd.DataFrame([[date, category, amount, description]], columns=['日期', '類別', '金額', '說明'])
            df = pd.concat([df, new_entry], ignore_index=True)

            df.to_csv('accounting.csv', index=False)
            print("記帳已儲存.")

            another_entry = input("還有帳要記嗎？ (是/否): ")
            if another_entry.lower() != '是':
                break

        elif action == '2' or action == '查詢':
            print("這是你的記帳紀錄:")
            display(df)
            break

        elif action == '3' or action == '返回':
            if not df.empty:
                df = df.iloc[:-1]
                df.to_csv('accounting.csv', index=False)
                print("你是不是哪裡記錯了?")
                print("(已移除最後一筆記帳紀錄.)")
            else:
                print("目前沒有記帳紀錄可以返回.")
            break


        else:
            print("\n是輸入 '記一筆', '查詢' 或 '返回'，或著輸入1，2，3也可以喔.")


def number_bomb():
    print("\n好诶，我們來玩數字炸彈！")
    while True:
        try:
            lower_bound = int(input("\n請輸入數字範圍的下限: "))
            upper_bound = int(input("\n請輸入數字範圍的上限: "))
            if lower_bound >= upper_bound:
                print("下限必須小於上限，請重新輸入。")
                continue
            target_number = random.randint(lower_bound, upper_bound)
            print(f"數字炸彈已設定在 {lower_bound} 到 {upper_bound} 之間。")
            guesses = 0

            while True:
                guess_input = input(f"請輸入你的猜測數字 ({lower_bound} 到 {upper_bound}): ")
                if guess_input == '不玩了':
                    print("\n好吧，隨時歡迎再來玩喲！")
                    return

                try:
                    guess = int(guess_input)
                    guesses += 1
                    if guess < lower_bound or guess > upper_bound:
                        print(f"請猜一個在 {lower_bound} 到 {upper_bound} 之間的數字。")
                    elif guess < target_number:
                        lower_bound = guess
                        print(f"太小了！從 {lower_bound} 到 {upper_bound}再選一個數字吧。")
                    elif guess > target_number:
                        upper_bound = guess
                        print(f"太大了！從 {lower_bound} 到 {upper_bound}再選一個數字吧。")
                    else:
                        print(f"恭喜中獎啦！你猜對了！數字是 {target_number}。")
                        print(f"話說獎賞是什麼")
                        break
                except ValueError:
                    print("好像哪裡不太對，請輸入數字。")

            play_again = input("還要再玩一次嗎？ (是/否): ")
            if play_again.lower() != '是':
                break
        except ValueError:
            print("\n請輸入有效的數字範圍。")

def meal_suggestion():
    meals = {
        '早': ['饅頭', '油條', '蛋餅', '稀飯'],
        '午': ['飯', '麵', '鹹酥雞', 'pizza'],
        '晚': ['泡麵', '7-11', '夜市', '不吃']
    }

    while True:
        meal_type = input("\n請選擇哪一餐 (早/午/晚): ")
        if meal_type in meals:
            break
        else:
            print("\n這是什麼意思，給我輸入 '早', '午', 或 '晚'.")

    suggestions = meals[meal_type]
    attempts = 0

    while attempts < 3:
        suggestion = random.choice(suggestions)
        print(f"那麼我建議你吃: {suggestion}")
        choice = input("如何，這個建議你喜歡嗎？ (是/否): ")
        if choice.lower() == '是':
            print("好的，祝您用餐愉快！")
            return
        else:
            attempts += 1

    print("\n我不行了，自己選擇啦!挑那麼九!")
    return


def main():
    """Main function to start the AI."""
    while True:
        print("\n需要我幫你做甚麼呢?")
        print("1. 記帳")
        print("2. 小遊戲")
        print("3. 餐點")
        print("4. 結束作業")

        choice = input("請輸入您的選擇 (1/2/3/4): ")

        if choice == '1' or choice == '記帳':
            accounting()
        elif choice == '2' or choice == '小遊戲':
            number_bomb()
        elif choice == '3' or choice == '餐點':
            meal_suggestion()
        elif choice == '4' or choice == '結束作業':
            print("好的，希望可以再次見到你")
            break
        else:
            print("\n錯瞜，沒有這個選項(可能後面增加吧...)，請重新輸入。。")

if __name__ == "__main__":
    main()
```
